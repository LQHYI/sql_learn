```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
```

```python
###Use pandas plot
df.plot(x = "Year", y = "Agriculture")
plt.xlabel("Year")
plt.ylabel("Percent")
plt.title("Percent of American women earn Agriculture's degree")
plt.show()
```

```python
###Use matplotlib.pyplot
df_year = df["Year"]
df_Agriculture = df["Agriculture"]
plt.plot(df_year, df_Agriculture,"-", color = "r", linewidth = 3)
plt.xlabel("Year")
plt.ylabel("Agriculture")
plt.title("Percentage of American women earn Agriculture's degree")
```

```python
df_2 = pd.read_csv("iris.csv")
df_2.columns = ['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'species']
df_2.plot(x = "sepal_len", y = "sepal_wid", kind ="scatter")
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Iris sepal length and width analysis")
```

```python
plt.plot(df_2["sepal_len"], df_2["sepal_wid"], "+", color = "r")
plt.xlabel("sepal length")
plt.xlabel("sepal width")
plt.title("Iris spel length and width analysis")
```

```python
df_2.describe()
```

```python
df_gbsp = df_2.groupby("species")["sepal_len"].agg(["count"])
```

```python
df_gbsp
```

```python
df_gbsp.plot(kind = "pie", autopct = '%.2f%%', fontsize = 10, figsize = (4,4), subplots = True, legend = False)
```

```python
df = pd.read_csv("iris.csv")
df.columns = ['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'species']
#df["species"] = df["species"].apply(lambda x: x.replace("Iris-"," "))
df_gbsp = df.groupby("species", as_index = False)["sepal_len"].agg({"counts":"count"})
plt.pie(df_gbsp["counts"], labels = df_gbsp["species"], autopct = "%.2f%%")
```

```python
df_gbsp = df.groupby("species", as_index = False).mean()
```

```python
df_gbsp.plot(kind = "bar")
plt.xticks(df_gbsp.index, df_gbsp["species"], rotation = 360)
```

```python
df_gbsp.plot(kind = "barh")
plt.yticks(df_gbsp.index, df_gbsp["species"], rotation = 360)
```

```python
df_gbsp.plot(kind = "bar", stacked = True)
plt.xticks(df_gbsp.index, df_gbsp["species"], rotation = 360)
```
